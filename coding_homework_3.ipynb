```python
#import np etc

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#import data as numpy arrats

v_rest_sst = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv')
v_rest_pvalb = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv')
```

```python
# calculate means

sst_mean = np.mean(v_rest_sst)
pvalb_mean = np.mean(v_rest_pvalb)

print(sst_mean)
print(pvalb_mean)
```

```python
#ttest, stat sig since pvalue > 0.05

t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(t_stat)
print(p_val)
```

```python
# plot histogram, if pvalue is sig plot both, if not print not sig, add label

p_value=stats.ttest_ind(v_rest_sst, v_rest_pvalb).pvalue


if p_value < 0.05:
  fig, ax = plt.subplots(1, 2)

  ax[0].hist(v_rest_sst)
  ax[0].set_title('sst')
  ax[0].set_xlabel('Voltage')
  ax[0].set_ylabel('Count')

  ax[1].hist(v_rest_pvalb)
  ax[1].set_title('pvalb')
  ax[1].set_xlabel('Voltage')
  ax[1].set_ylabel('Count')

else:
  print("p value is not sig")
```

```python
#define vresthist

def vrestHist(v_rest_sst, v_rest_pvalb):
    """
    function to make hist of v_rest_sst and v_rest_pvalb IF pvalue is sig
    """
    p_value = stats.ttest_ind(v_rest_sst, v_rest_pvalb).pvalue

    if p_value < 0.05:
        fig, ax = plt.subplots(1, 2, figsize=(12, 5))

        ax[0].hist(v_rest_sst)
        ax[0].set_title('sst')
        ax[0].set_xlabel('Voltage')
        ax[0].set_ylabel('Count')

        ax[1].hist(v_rest_pvalb)
        ax[1].set_title('pvalb')
        ax[1].set_xlabel('Voltage')
        ax[1].set_ylabel('Count')

    else:
        print("p value is not sig")

    plt.show()
```

```python
#calls function

vrestHist(v_rest_sst, v_rest_pvalb)
```

```python
# scatter plot with color code

for i in range(len(v_rest_sst)):
  if v_rest_sst[i] > -60 and v_rest_pvalb[i] > -60:
    color = 'red'
  elif v_rest_sst[i] > -60 and v_rest_pvalb[i] <= -60:
    color = 'green'
  elif v_rest_sst[i] <= -60 and v_rest_pvalb[i] <= -60:
    color = 'blue'
  elif v_rest_sst[i] <= -60 and v_rest_pvalb[i] > -60:
    color = 'purple'

  plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color=color)

plt.xlabel('v_rest_sst')
plt.ylabel('v_rest_pvalb')
plt.show()
```

```python
#define vrestscatter

def vrestScatter(v_rest_sst, v_rest_pvalb):
  """
  function to make color coded scatterplot based on resting voltage of neurons
  """
  for i in range(len(v_rest_sst)):
    if v_rest_sst[i] > -60 and v_rest_pvalb[i] > -60:
      color = 'red'
    elif v_rest_sst[i] > -60 and v_rest_pvalb[i] <= -60:
      color = 'green'
    elif v_rest_sst[i] <= -60 and v_rest_pvalb[i] <= -60:
      color = 'blue'
    elif v_rest_sst[i] <= -60 and v_rest_pvalb[i] > -60:
      color = 'purple'

    plt.scatter(v_rest_sst[i], v_rest_pvalb[i], color=color)

  plt.xlabel('v_rest_sst')
  plt.ylabel('v_rest_pvalb')
  plt.show()
```

```python
#call vrestScatter

vrestScatter(v_rest_sst, v_rest_pvalb)
```
