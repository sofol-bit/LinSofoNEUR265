```python
#2 import modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 imports data

acc_hits = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_hits_hw5.csv', delimiter = ',')
acc_fa = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/acc_fa_hw5.csv', delimiter = ',')
```

```python
#4 print shapes

print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#6 create time variable

time = np.linspace(0, 4, 8001)
```

```python
#7 sampling rate

8001/4
```

```python
#8 subplots

fig, ax = plt.subplots(2, 1, figsize=(10, 6))

ax[0].plot(time,acc_hits[0])
ax[0].set_ylabel('Voltage')
ax[0].set_xlabel('Time')
ax[0].set_title('Acc Hits')

ax[1].plot(time,acc_fa[0])
ax[1].set_ylabel('Voltage')
ax[1].set_xlabel('Time')
ax[1].set_title('Acc False Alarms')
```

hits: 2.5 sec to 4 sec, 6 peaks... about 4 Hz

false alarm: 2 sec to 3.5 sec, 3 peaks... about 2 Hz

false alarm is slower and less rythmic than hits

```python
#10 ERP hits

erp_hits = np.mean(acc_hits, axis=0)

plt.figure(figsize=(8, 5))
plt.plot(time, erp_hits)
plt.axvline(x=2, color='black')
plt.axhline(y=0, color='black')
plt.xlabel('Time')
plt.ylabel('Voltage')
```

Seems to start reacting right at screen touch, almost flat prior

```python
#12 FA ERP

erp_hits = np.mean(acc_fa, axis=0)

plt.figure(figsize=(8, 5))
plt.plot(time, erp_hits)
plt.axvline(x=2, color='black')
plt.axhline(y=0, color='black')
plt.xlabel('Time')
plt.ylabel('Voltage')
```

no consistent response relative to hits
