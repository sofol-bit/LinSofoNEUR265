```python
#import np, plt, st

import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
#import csv as arrays

v_rest1 = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv')

v_rest2 = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv')
```

```python
#concatenating each list, tried concatenate but it gave shape of 1000, so I used vstack
#check that is is (2,500)

v_rest = np.vstack([v_rest1, v_rest2])
print(v_rest.shape)
```

```python
#check size of v_rest

print(v_rest.size)
```

```python
# print first row

v_rest[0]
```

```python
#sum values in third column

v_rest[:,2].sum()
```

```python
#second to last value of the second row, neg indexing

v_rest[1,-2]
```

```python
#create histogram with axis labels

plt.hist(v_rest[0,:])
plt.xlabel('Voltage')
plt.ylabel('Count')
```

```python
#means of each row

v_rest_mean_sst = v_rest[0,:].mean()
print(v_rest_mean_sst)

v_rest_mean_pvalb = v_rest[1,:].mean()
print(v_rest_mean_pvalb)
```
